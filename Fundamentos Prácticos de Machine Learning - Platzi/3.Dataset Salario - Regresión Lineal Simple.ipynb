```python
#VIDEO 11 - REGRESIÓN LINEAL SIMPLE CON SCIKIT LEARN
import pandas as pd #Librería pandas y la llama pd
import matplotlib.pyplot as plt #Para verlo visualmente y lo llama plt
from sklearn.model_selection import train_test_split #Para dividir datos
from sklearn.linear_model import LinearRegression #Importa el algoritmo 
```

```python
salary=pd.read_csv('Salary_Data.csv') #Acá importo el dataset.
salary.head(10) 
```

```python
#VIDEO 12
salary.shape #Para ver cuántos datos tiene. Tiene 30 ejemplos en 2 columnas
```

```python
x = salary.iloc[:, :-1].values
y = salary.iloc[:, 1].values
```

```python
x_train_x, x_test_x, y_train_y, y_test_y = train_test_split(x,y,test_size = 0.20, random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(x_train_x, y_train_y)
```

```python
viz_train=plt #Muestra qué aprendió el modelo, lo visualiza de manera gráfica
viz_train.scatter(x_train_x, y_train_y, color = 'blue')
viz_train.plot(x_train_x, regressor.predict(x_train_x), color = 'black')
viz_train.title('Salario vs Experiencia')
viz_train.xlabel('Experiencia')
viz_train.ylabel('Salario')
viz_train.show()
```

```python
viz_train=plt #Muestra qué aprendió el modelo, lo visualiza de manera gráfica
viz_train.scatter(x_test_x, y_test_y, color = 'red')
viz_train.plot(x_train_x, regressor.predict(x_train_x), color = 'black')
viz_train.title('Salario vs Experiencia')
viz_train.xlabel('Experiencia')
viz_train.ylabel('Salario')
viz_train.show()
```

```python
regressor.score(x_test_x, y_test_y)
```
